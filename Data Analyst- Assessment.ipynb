```python
# Converting and uploading Receipts data from json file
f = open(r'receipts.json')
rdata = [json.loads(line) for line in f]  
```

```python
receipts_dff = pd.json_normalize(rdata)
receipts_dff.head(2) 
```

```python
receipts_df = pd.json_normalize(rdata)
receipts_df.head(2) 
```

```python
# translating the data which is in format 'epoch' to 'datetime'
receipts_df['createDate.$date'] = pd.to_datetime(receipts_df['createDate.$date']).dt.date
receipts_df['dateScanned.$date'] = pd.to_datetime(receipts_df['dateScanned.$date']).dt.date
receipts_df['finishedDate.$date'] = pd.to_datetime(receipts_df['finishedDate.$date']).dt.date
receipts_df['modifyDate.$date'] = pd.to_datetime(receipts_df['modifyDate.$date']).dt.date
receipts_df['pointsAwardedDate.$date'] = pd.to_datetime(receipts_df['pointsAwardedDate.$date']).dt.date
receipts_df['purchaseDate.$date'] = pd.to_datetime(receipts_df['purchaseDate.$date']).dt.date

receipts_df.head() 
```

```python
# Replacing null values with nan
receipts_df['rewardsReceiptItemList'] = receipts_df['rewardsReceiptItemList'].replace(np.nan, 0)
```

```python
#In dataframe'receipts_df' converting the column 'rewardsReceiptItemList' which is nested json as new dataframe'rewardsReceipt_df', 
#Foreign key 'receipt_ids' is created to connect two tables 'receipts_df' and 'rewardsReceipt_df'

rewardsReceipt_df = pd.DataFrame()
count = 1
receipt_ids = []
data = []
for i in range(len(receipts_df.rewardsReceiptItemList)):
    if receipts_df.rewardsReceiptItemList[i] != 0:
        for j in receipts_df.rewardsReceiptItemList[i]:
            j['receipt_id'] = str(count)
            data.append(j)
        receipt_ids.append(j['receipt_id'])
        count += 1
    else:
        receipt_ids.append('0')
```

```python
rewardsReceipt_df = pd.DataFrame.from_dict(data)
```

```python
receipts_df['receipt_ids'] = receipt_ids
```
